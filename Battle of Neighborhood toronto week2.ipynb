# Data 

#### Pick up the useful data and classify them

```python
Transportation=['Airport Lounge','Bus Stop','Light Rail Station','Lounge','Trail']
Scenic_spot=['Art Gallery','Church','Concert Hall','Construction & Landscaping','Fountain','Garden','History Museum',
            'Lake','Museum','Outdoor Sculpture','Park','Performing Arts Venue','Plaza','Scenic Lookout']
Food=['American Restaurant','Asian Restaurant','BBQ Joint','Bakery','Belgian Restaurant',
      'Breakfast Spot','Bubble Tea Shop','Burger Joint','Burrito Place','Café','Candy Store','Caribbean Restaurant',
     'Coffee Shop','Colombian Restaurant','Comfort Food Restaurant','Cuban Restaurant','Dessert Shop','Diner',
     'Ethiopian Restaurant','Falafel Restaurant','Fast Food Restaurant','Food','Food & Drink Shop','Food Court','Food Truck',
     'French Restaurant','Gluten-free Restaurant','Greek Restaurant','Ice Cream Shop','Indian Restaurant','Italian Restaurant',
     'Japanese Restaurant','Korean Restaurant','Latin American Restaurant','Malay Restaurant','Mexican Restaurant',
     'Middle Eastern Restaurant','New American Restaurant','Noodle House','Pizza Place','Ramen Restaurant','Restaurant',
     'Seafood Restaurant','Steakhouse','Sushi Restaurant','Taco Place','Thai Restaurant','Theme Restaurant',
     'Vegetarian / Vegan Restaurant']
Hotel=['Hostel','Hotel','Pub']
Shop=['Adult Boutique','Clothing Store','Costume Shop','Fish & Chips Shop','Grocery Store','Supermarket',
     'Toy / Game Store']
Entertainment=['Bar','Beer Bar','Cocktail Bar','Dance Studio','Dive Bar','Gastropub','Gay Bar','General Entertainment',
              'Gym','Hotel Bar','Jazz Club','Movie Theater','Sports Bar','Theater','Wine Bar']
```

```python
toronto_venues_temp=toronto_venues
tot_class=Transportation+Scenic_spot+Food+Hotel+Shop+Entertainment
toronto_venues_temp=toronto_venues_temp[toronto_venues_temp['Venue Category'].isin(tot_class)]
print(toronto_venues_temp.shape)
toronto_venues_temp.head()
```

#### The unique venue categories reduced from 150 to 93

```python
print('There are {} uniques categories.'.format(len(toronto_venues_temp['Venue Category'].unique())))
```

#### Classify the new venue categories

```python
toronto_category=toronto_venues_temp[:]
```

```python

toronto_category['Venue Category'].replace(Food,'Food')
toronto_category['Venue Category'].replace(Transportation,'Transportation')
toronto_category['Venue Category'].replace(Scenic_spot,'Scenic_spot')
toronto_category['Venue Category'].replace(Entertainment,'Entertainment')
toronto_category['Venue Category'].replace(Shop,'Shop')
toronto_category['Venue Category'].replace(Hotel,'Hotel')
print(toronto_category.shape)
toronto_category.head()
```

```python
toronto_venues_temp.head()
```

```python
# one hot encoding
toronto_category_onehot = pd.get_dummies(toronto_category[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
toronto_category_onehot['Neighbourhood'] = toronto_category['Neighbourhood'] 

# move neighborhood column to the first column
fixed_columns = [toronto_category_onehot.columns[-1]] + list(toronto_category_onehot.columns[:-1])
toronto_category_onehot = toronto_category_onehot[fixed_columns]
print(toronto_category_onehot.shape)
toronto_category_onehot.head()
```

#### Now, the neighbourhood data is very clear and succinct for new visitor

# Methodology

#### Group rows by neighbourhood and by taking the mean of the frequency of occurrence of each category

```python
import numpy as np # library to handle data in a vectorized manner
import pandas as pd # library for data analsysis

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

import json # library to handle JSON files


from geopy.geocoders import Nominatim  #convert an address into latitude and longitude values

import requests # library to handle requests
from pandas.io.json import json_normalize # tranform JSON file into a pandas dataframe

# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors

# import k-means from clustering stage
from sklearn.cluster import KMeans


import folium # map rendering library

print('Libraries imported.')
```

```python
toronto_grouped = toronto_category_onehot.groupby('Neighbourhood').mean().reset_index()
print(toronto_grouped.shape)
toronto_grouped.head()        # check grouped data
```

#### Print each neighbourhood along with all venues

```python
num_top_venues = 6    # Now, there are total 6 venues
for neigh in toronto_grouped['Neighbourhood']:
    print("----"+neigh+"----")
    temp = toronto_grouped[toronto_grouped['Neighbourhood'] == neigh].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

#### Sort the venues in descending order

```python
# function to sort the venues in descending order
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
num_top_venues = 6

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighbourhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighbourhood'] = toronto_grouped['Neighbourhood']

for ind in np.arange(toronto_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(toronto_grouped.iloc[ind, :], num_top_venues)

print(neighborhoods_venues_sorted.shape)
neighborhoods_venues_sorted.head()   # check sorted venues data
```

#### Cluster Neighbourhoods

```python
# set number of clusters
kclusters = 5

toronto_grouped_clustering = toronto_grouped.drop('Neighbourhood', 1)
# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=1).fit(toronto_grouped_clustering)

# check cluster labels generated for each row in the dataframe
print(kmeans.labels_[0:10])
print(len(kmeans.labels_))
```

#### Merge data

```python
Toronto_Dict_csv = 'Toronto Dict.csv'   
toronto_dict = pd.read_csv(Toronto_Dict_csv)    # read Toronto Postcode and geospatial data
print(toronto_dict.shape)
toronto_dict.head()  # check postcode data
```

```python
toronto_data = toronto_dict[toronto_dict['Borough'].str.contains("Toronto")].reset_index(drop=True)
print(toronto_data.shape)
toronto_data.head()
```

```python
geocsv_data = pd.read_csv('toronto_m.geospatial_data.csv')
print(geocsv_data.shape)
geocsv_data.head()       # check geo data
```

```python
toronto_di=neighborhoods_venues_sorted.join(geocsv_data)  # merge Toronto sorted venues and geospatial data
print(toronto_di.shape)
toronto_di.head()
```

```python
toronto_merged = toronto_di
toronto_merged['Cluster Labels'] = kmeans.labels_    # add clustering labels

# merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
toronto_merged=toronto_merged.reindex(columns=['Neighbourhood','Postal Code','Latitude','Longitude','Cluster Labels','1st Most Common Venue',
                                '2nd Most Common Venue','3rd Most Common Venue','4th Most Common Venue','5th Most Common Venue',
                               '6th Most Common Venue'])

toronto_merged.head()  
```

```python
map_clusters = folium.Map(location=[43.784535, -79.194353], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i+x+(i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(toronto_merged['Latitude'], toronto_merged['Longitude'], toronto_merged['Neighbourhood'],kmeans.labels_):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
